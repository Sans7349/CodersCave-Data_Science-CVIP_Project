```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
file_path = '/content/Hackathon_Working_Data.csv'
data = pd.read_csv(file_path)
purchase_frequency = data['BILL_ID'].value_counts()
product_popularity = data['BRD'].value_counts().head(10)
average_spending_per_transaction = data['BILL_AMT'].mean()
total_spending_per_customer = data.groupby('BILL_ID')['BILL_AMT'].sum().mean()
transactions_per_month = data['MONTH'].value_counts()
transactions_per_day = data['DAY'].value_counts()
numerical_columns = data.select_dtypes(include=['float64', 'int64']).columns
correlation_matrix = data[numerical_columns].corr()
sns.set(style="whitegrid")

# 1. Distribution of Purchase Frequency
plt.figure(figsize=(10, 6))
sns.histplot(purchase_frequency, bins=30, kde=False, color='blue')
plt.title('Distribution of Purchase Frequency')
plt.xlabel('Number of Purchases per Customer')
plt.ylabel('Frequency')
plt.show()

# 2. Top 10 Most Common Products
plt.figure(figsize=(10, 6))
sns.barplot(x=product_popularity.values, y=product_popularity.index, palette='viridis')
plt.title('Top 10 Most Common Products')
plt.xlabel('Number of Purchases')
plt.ylabel('Product')
plt.show()

# 3. Monthly Transactions
plt.figure(figsize=(10, 6))
sns.barplot(x=transactions_per_month.index, y=transactions_per_month.values, palette='viridis')
plt.title('Transactions per Month')
plt.xlabel('Month')
plt.ylabel('Number of Transactions')
plt.show()

# 4. Daily Transactions
plt.figure(figsize=(10, 6))
sns.barplot(x=transactions_per_day.index, y=transactions_per_day.values, palette='viridis')
plt.title('Transactions per Day')
plt.xlabel('Day')
plt.ylabel('Number of Transactions')
plt.show()

# 5. Spending Patterns
plt.figure(figsize=(10, 6))
sns.histplot(data['BILL_AMT'], bins=30, kde=False, color='blue')
plt.title('Distribution of Spending per Transaction')
plt.xlabel('Spending per Transaction')
plt.ylabel('Frequency')
plt.show()

# 6. Correlation Heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix')
plt.show()
```
